Missing Values imputation using Scikit learn for Numeric and Categorical variuable

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

```python
train = pd.read_csv('train.csv')

print("shape of train df = ",train.shape)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
train.head()
```

```python
X_train = train.drop(columns='SalePrice')
y_train = train['SalePrice']
```

```python
print("shape of X_train df = ",X_train.shape)
print("shape of y_train df = ",y_train.shape)
```

Numerical missing value impution

```python
n_vars = X_train.select_dtypes(include=['int64','float64']).columns
n_vars
```

```python
X_train[n_vars].isnull().sum()
```

```python
imputer_mean = SimpleImputer(strategy='mean')
#imputer_mean = SimpleImputer(strategy='constant', fill_value=99)
```

```python
imputer_mean.fit(X_train[n_vars])
```

```python
imputer_mean.statistics_
```

```python
imputer_mean.transform(X_train[n_vars])
```

```python
X_train[n_vars] = imputer_mean.transform(X_train[n_vars])
```

```python
X_train[n_vars].isnull().sum()
```

Catogrical missing value Impution

```python
cat_vars = X_train.select_dtypes(include=['O']).columns
cat_vars
```

```python
X_train[cat_vars].isnull().sum()
```

```python
imputer_mode = SimpleImputer(strategy='most_frequent')
#imputer_mean = SimpleImputer(strategy='constant', fill_value=99)
imputer_mode
```

```python
imputer_mode.fit(X_train[cat_vars])
```

```python
imputer_mode.statistics_
```

```python
X_train[cat_vars] = imputer_mode.transform(X_train[cat_vars])
```

```python
X_train[cat_vars].isnull().sum()
```
