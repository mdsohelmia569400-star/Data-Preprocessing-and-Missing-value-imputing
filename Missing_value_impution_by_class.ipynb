```python
missing_variable_per = df.isnull().sum()/df.shape[0]*100
missing_variable_per
```

```python
Misssing_variable_col = missing_variable_per[missing_variable_per>20].keys()
Misssing_variable_col
```

```python

df2_drop_columns = df.drop(Misssing_variable_col, axis=1)   # drop columns missing 20% or avobe null value columns
df2_drop_columns.shape
df2_drop_columns
```

```python
df2_drop_columns.shape
```

```python
d3_num = df2_drop_columns.select_dtypes(include=['int64', 'float64'])
d3_num.shape
```

```python
d3_num.isnull().sum()
```

```python
missing_num_var = [var for var in d3_num.columns if d3_num[var].isnull().mean()>0]
missing_num_var       ### show name all null value column
```

```python
num_var_miss = ['LotFrontage',
 'MasVnrArea',
 'BsmtFinSF1',
 'BsmtFinSF2',
 'BsmtUnfSF',
 'TotalBsmtSF',
 'BsmtFullBath',
 'BsmtHalfBath',
 'GarageYrBlt',
 'GarageCars',
 'GarageArea']

d3_num[num_var_miss][d3_num[num_var_miss].isnull().any(axis=1)]
```

```python
df['LotConfig'].unique()   ### Cheack classes
```

```python
df[df.loc[:,'LotConfig']=='Inside']['LotFrontage'].replace(np.nan, df[df.loc[:,'LotConfig']=='Inside']['LotFrontage'].mean())
```

```python
df_copy = df.copy()
for var_class in df ['LotConfig'].unique():
  df_copy.update(df[df.loc[:,'LotConfig']==var_class]['LotFrontage'].replace(np.nan, df[df.loc[:,'LotConfig']==var_class]['LotFrontage'].mean()))
```

```python
df_copy.isnull().sum()
```

```python
df_copy = df.copy()
num_vars_miss = ['LotFrontage','MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','BsmtFullBath','BsmtHalfBath','GarageYrBlt','GarageCars','GarageArea']
cat_vars = ['LotConfig','MasVnrType','MSZoning','Utilities','Exterior1st','Exterior2nd','KitchenQual','Functional','SaleType']
for cat_var, num_var_miss in zip(cat_vars,num_var_miss):
  for var_class in df ['LotConfig'].unique():
    df_copy.update(df[df.loc[:,'LotConfig']==var_class][num_vars_miss].replace(np.nan, df[df.loc[:,'LotConfig']==var_class][num_vars_miss].mean()))
```

```python
plt.figure(figsize=(10,8))
sns.set()
for i, var in enumerate(missing_num_var):
  plt.subplot(4,4,i+1)
  sns.distplot(df[var],bins=20, label = 'Orginal')
  sns.displot(df_copy[var], bins = 20, label = 'Mean')                  ### Show data distribution
  plt.tight_layout()
```
